```python
# 🚀 Ensure GPU is available
import torch
if not torch.cuda.is_available():
    raise SystemError("❌ No NVIDIA GPU found. Please enable GPU in Colab (Runtime > Change runtime type > GPU) and restart the notebook.")

# 🚀 Install required libraries
!pip install diffusers transformers accelerate torch matplotlib --quiet

# 📌 Import necessary libraries
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt

# ✅ Load Stable Diffusion Model from Hugging Face
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5").to("cuda")

# 🔹 Function to Generate and Display Images
def generate_image(prompt):
    print(f"🖼 Generating image for: {prompt}")
    image = pipe(prompt).images[0]

    # 📌 Display the generated image
    plt.figure(figsize=(8, 8))
    plt.imshow(image)
    plt.axis("off")
    plt.show()

# 🔥 Example Prompts
prompts = [
    "A futuristic city skyline at sunset, cyberpunk style",
    "An astronaut riding a horse on Mars",
    "A magical forest with glowing mushrooms and fairies"
]

# 🚀 Generate Images for All Prompts
for prompt in prompts:
    generate_image(prompt)
```
